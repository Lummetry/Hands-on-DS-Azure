```
df_diabetes = df_diabetes.iloc[:, 1:]
```

```
df_X = df_diabetes.iloc[:,:-1]
```

```
df_X.iloc[:10]
```

```
X = df_X.values
y = ser_y.values
```

```
df_X.describe()
```

```
X = (X - X.mean()) / X.std()
```

```
pd.DataFrame(data=X).describe()
```

```
X.shape[1]
```

```

tf_input = tf.keras.layers.Input((X.shape[1],), name='input_layer')
hidden_layer = tf.keras.layers.Dense(20, name='hidden_linearity')
hidded_activation = tf.keras.layers.Activation('relu', name='hidden_activation')
regression_layer = tf.keras.layers.Dense(1, name='readout_linearity')
tf_out = regression_layer(hidded_activation(hidden_layer(tf_input)))

model = tf.keras.models.Model(tf_input, tf_out)
opt = tf.keras.optimizers.Adam(lr=0.01)
loss = tf.keras.losses.MSE

model.compile(loss=loss, optimizer=opt, metrics=['mae'])
```

```
from sklearn.model_selection import train_test_split

x_trn, x_test, y_trn, y_test = train_test_split(X,y, test_size=0.1)
print(x_trn.shape)
print(x_test.shape)
```

```
x_train, x_dev, y_train, y_dev = train_test_split(x_trn, y_trn, test_size=0.1)
print(x_train.shape)
print(x_dev.shape)
```

```
hist = model.fit(x_train, y_train, validation_data=(x_dev, y_dev), epochs=50, batch_size=32)
test_MAE = model.evaluate(x_test, y_test)[1]
```

```
plt.figure(figsize=(10,6))
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.legend(['train_loss', 'dev_loss'])
plt.title('Train vs dev loss progression')
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.plot(hist.history['mean_absolute_error'])
plt.plot(hist.history['val_mean_absolute_error'])
plt.legend(['train_MAE', 'dev_MAE'])
plt.title('Train vs dev MAE progression')
plt.axhline(test_MAE, color='r')
plt.show()
```

```
import azureml
from azureml.core import Workspace

workspace_name = 'hodsws'
workspace_location = 'West Europe'
resource_group = 'ML'
subscription_id = dbutils.secrets.get(scope='config',key='subscription')

workspace = Workspace.create(name = workspace_name,
                             location = workspace_location,
                             resource_group = resource_group,
                             subscription_id = subscription_id,
                             exist_ok=True)
```

```
workspace.get_details()
```

```
import mlflow
import mlflow.keras

with mlflow.start_run() as active_run:
  mlflow.keras.log_model(add_model, 'hodsaddmodel')
  #mlflow.keras.save_model(add_model, path='hodsaddmodel')
  run_id = active_run.info.run_id

model_uri = "runs:/" + run_id + "/hodsaddmodel"
print(model_uri)
```

```
loaded_model = mlflow.keras.load_model(model_uri=model_uri)
loaded_model.summary()
loaded_model.predict([100,1000])
```

```
import mlflow.azureml as mlfaz

model_image, azure_model = mlfaz.build_image(
  model_uri=model_uri,
  workspace=workspace,
  model_name='hodsaddmodel',
  image_name='hodsaddmodelimage',
  description='Dummy tf.keras model +99',
  synchronous=False,
)

model_image.wait_for_creation(show_output=True)
```

```
from azureml.core.webservice import AciWebservice, Webservice

api_name = 'hods-add-99-model'
api_config = AciWebservice.deploy_configuration()

try:
  webservice = Webservice.deploy_from_image(
    name=api_name, 
    image=model_image,
    deployment_config=api_config,
    workspace=workspace)
  print("Creation succesfull')
except:
  pass
```
